### Procedure: 
Using the networkxgraph package, I created a graph called G. From there, I added in the nodes, which are the names and then added in the edges which connect the names. Once this was complete, I then used the matplotlib package to plot the graph, G, which shows all the nodes (names) connected via the edges.  


```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(['Carol', 'Andre', 'Fernando', 'Diane', 'Beverly', 'Garth', 'Ed', 'Heather', 'Ike', 'Jane'])
```

```python
G.add_edges_from([('Diane', 'Carol'), ('Diane', 'Andre'), ('Diane', 'Fernando'), ('Diane', 'Beverly'), ('Diane', 'Ed'), ('Diane', 'Garth'), ('Carol', 'Andre'),('Carol', 'Fernando'), ('Andre', 'Fernando'), ('Andre', 'Beverly'), ('Fernando', 'Garth'), ('Beverly', 'Garth'), ('Beverly', 'Ed'), ('Ed', 'Garth'), ('Fernando', 'Heather'), ('Garth', 'Heather'), ('Heather', 'Ike'), ('Ike', 'Jane')])
```

```python
nx.draw(G, with_labels=True)
```
