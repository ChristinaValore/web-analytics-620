```python
import networkx as nx
import matplotlib.pyplot as plt
from pylab import rcParams
import numpy as np
```

```python
G = nx.read_gml('word_adjacencies.gml')
```

```python
print(nx.info(G)) #info about the graph
print('Diameter:', nx.diameter(G)) #diameter
```

```python
nx.draw(G, with_labels=True) #Draw the graph 
```

```python
pos = nx.random_layout(G, seed=30)
nx.draw(G,pos=pos,with_labels=True) #Draw the graph 
```
